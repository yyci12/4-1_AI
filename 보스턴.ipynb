```python
import tensorflow as tf
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
# 학습 데이터 #

train= pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/train_dataset.csv')
test= pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/test_dataset.csv')
print(train.columns)
train.head()
```

```python
ind = train[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
sub = train[['MEDV']]
print(ind.shape, sub.shape)
```

```python
X = tf.keras.layers.Input(shape=[13])
Y = tf.keras.layers.Dense(1)(X)
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse')
```

```python
model.fit(ind, sub, epochs=1000, verbose=0)
model.fit(ind, sub, epochs=10)
```

```python
print(model.predict(ind[5:10]))
```

```python
print(sub[5:10])
```

```python
print(model.get_weights())
```

```python
pred = model.predict(test)
print(pred)
print(type(pred))
dataframe = pd.DataFrame(pred)
print(type(dataframe))
```
